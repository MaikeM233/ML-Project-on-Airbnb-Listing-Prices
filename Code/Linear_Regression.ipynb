```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
# scikit-learn

from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge, RidgeCV, Lasso, LassoCV
```

```python
train_dummy_scale = pd.read_excel('train_dummy.xlsx', index_col=0)
test_dummy_scale = pd.read_excel('test_dummy.xlsx', index_col=0)
```

## Linear Regression

```python
train = train_dummy_scale.iloc[:,2:]
train.head()
```

```python
test = test_dummy_scale.iloc[:,2:]
test.head()
```

```python
result = sm.OLS(train.price, train.drop('price', axis=1)).fit()
result.summary()
```

```python
X_train = train.drop(['price'], axis = 1)
X_test = test.drop(['price'], axis = 1)
y_train = train['price']
y_test = test['price']
```

```python
linear = LinearRegression()
linear.fit(X_train,y_train)
```

```python
y_pred = linear.predict(X_test)
```

```python
# Mean Squared Error
ols_mse = mean_squared_error(y_pred, y_test)
print('ols_mse', ols_mse)
```

```python
ols_R2 = linear.score(X_train, y_train)
print('ols_R2', ols_R2)
```

### Ridge

```python
lbd = np.logspace(-10, 3, 100)
```

```python
# Choose Î» by cross-validation
ridgecv = RidgeCV(alphas = lbd, cv = 10)
ridgecv.fit(X_train, y_train)
ridgecv.alpha_
```

```python
ridge = Ridge(alpha = ridgecv.alpha_)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
ridge_mse = mean_squared_error(y_test, y_pred)
print('ridge_mse', ridge_mse)
```

```python
ridge_R2 = ridge.score(X_train, y_train)
print('ridge_R2', ridge_R2)
```

### Lasso

```python
lbd = np.logspace(-5, 3, 100)
```

```python
# Choose Î» by cross-validation
lassocv = LassoCV(alphas = lbd, cv = 10)
lassocv.fit(X_train, y_train)
lassocv.alpha_
```

```python
lasso = Lasso(alpha = lassocv.alpha_)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)
lasso_mse = mean_squared_error(y_test, y_pred)
print('lasso_mse', lasso_mse)
```

```python
lasso_R2 = lasso.score(X_train, y_train)
print('lasso_R2', lasso_R2)
```
